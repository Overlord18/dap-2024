```python
#1
class Solution:
    def twoSum(self, nums, target):
        # Инициализация хеш-таблицы для хранения значений и их индексов
        num_dict = {}

        # Проход по каждому числу в массиве
        for index, num in enumerate(nums):
            # Вычисляем число, необходимое для достижения целевой суммы
            complement = target - num

            # Проверяем, находится ли необходимое число в хеш-таблице
            if complement in num_dict:
                # Возвращаем индексы текущего числа и найденного числа
                return [num_dict[complement], index]

            # Если не найдено, добавляем текущее число в хеш-таблицу
            num_dict[num] = index

        # Если решения нет, возвращаем пустой список
        return []

# Примеры использования
sol = Solution()
result_1 = sol.twoSum([2, 67, 11, 5], 78)
print(f"Исходная последовательность: {tuple([2, 67, 11, 5])}, Суммы двух чисел: 78, Индексы чисел: {result_1}")

result_2 = sol.twoSum([3, 2, 4], 6)
print(f"Исходная последовательность: {tuple([3, 2, 4])}, Суммы двух чисел: 6, Индексы чисел: {result_2}")

result_3 = sol.twoSum([3, 3], 6)
print(f"Исходная последовательность: {tuple([3, 3])}, Суммы двух чисел: 6, Индексы чисел: {result_3}")
```

```python
#2
def isPalindrome(x):
    # Отрицательные числа не могут быть палиндромами
    if x < 0:
        result = False
    else:
        # Преобразуем число в строку
        str_x = str(x)

        # Проверяем, равно ли число своей обратной строке
        result = str_x == str_x[::-1]

    # Формируем строку с результатом для вывода
    result_str = (
        f"Исходное число: {x}. "
        f"Является палиндромом: {'Да' if result else 'Нет'}."
    )

    # Выводим результаты
    print(result_str)

    return result

# Примеры использования
isPalindrome(121)   # Вывод: Исходное число: 121. Является палиндромом: Да.
isPalindrome(-121)  # Вывод: Исходное число: -121. Является палиндромом: Нет.
isPalindrome(10)    # Вывод: Исходное число: 10. Является палиндромом: Нет.
```

```python
#3
def romanToInt(s):
    # Создаем словарь для преобразования римских цифр в целые числа
    roman_to_int = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }

    # Инициализируем переменную для накопления общего результата
    total = 0
    prev_value = 0

    # Обрабатываем символы справа налево
    for char in reversed(s):
        value = roman_to_int[char]
        # Если текущее значение меньше предыдущего, вычитаем его
        if value < prev_value:
            total -= value
        else:
            total += value
        prev_value = value

    return total

# Примеры использования
print(romanToInt("III"))      # Вывод: 3
print(romanToInt("IV"))       # Вывод: 4
print(romanToInt("IX"))       # Вывод: 9
print(romanToInt("LVIII"))    # Вывод: 58
print(romanToInt("MCMXCIV"))  # Вывод: 1994
```

```python
#4
def longestCommonPrefix(strs):
    # Проверяем, является ли список пустым
    if not strs or len(strs) == 0:
        return ""

    # Берём первый элемент списка как потенциальный префикс
    prefix = strs[0]

    # Перебираем остальные строки в списке
    for s in strs[1:]:
        # Пока текущее слово не начинается с текущего префикса
        while not s.startswith(prefix):
            # Сокращаем префикс
            prefix = prefix[:-1]
            # Если префикс стал пустым, дальше сравнивать нечего
            if not prefix:
                return ""

    return prefix

# Примеры использования
print(longestCommonPrefix(["flower", "flow", "flight"]))          # Вывод: fl
print(longestCommonPrefix(["dog", "racecar", "car"]))              # Вывод: (пустая строка)
print(longestCommonPrefix(["interspecies", "interstellar", "interstate"]))  # Вывод: inter
```

```python
#5
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def printList(head):
    # Функция для печати содержимого связанного списка
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

def mergeTwoLists(list1, list2):
    # Создаем фиктивный узел начала нового списка
    dummy = ListNode()
    current = dummy

    # Итерация по обоим спискам одновременно
    while list1 and list2:
        if list1.val <= list2.val:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next
        current = current.next

    # Присоединяем остаток оставшегося списка
    if list1:
        current.next = list1
    elif list2:
        current.next = list2

    return dummy.next

# Пример использования
if __name__ == "__main__":
    # Создание тестовых списков
    list1 = ListNode(2, ListNode(4, ListNode(5, ListNode(6))))
    list2 = ListNode(1, ListNode(2, ListNode(3, ListNode(5, ListNode(6)))))

    # Выполняем слияние списков
    merged_list = mergeTwoLists(list1, list2)

    # Печать исходных и результирующего списков
    print("Исходный список 1:", printList(list1))
    print("Исходный список 2:", printList(list2))
    print("Объединенный список:", printList(merged_list))
```

```python
#6
def lengthOfLongestSubstring(s):
    # Словарь для хранения последнего появления каждого символа
    char_map = {}
    start = 0
    max_length = 0

    # Проходим по строке
    for end in range(len(s)):
        char = s[end]

        # Если символ ранее встречался и находится в текущем окне
        if char in char_map and char_map[char] >= start:
            start = char_map[char] + 1

        # Обновляем позицию символа
        char_map[char] = end

        # Рассчитываем максимальную длину
        current_length = end - start + 1
        max_length = max(max_length, current_length)

    return max_length


def printLongestSubstringResult(s):
    # Находим длину самой длинной уникальной подстроки
    length = lengthOfLongestSubstring(s)

    # Формируем строку с результатом
    result_str = f"Исходная строка: {s}\nМаксимальная длина неповторяющейся подстроки: {length}"

    # Выводим результат
    print(result_str)

    return length


# Примеры использования
printLongestSubstringResult("abcabcbb")
printLongestSubstringResult("bbbbb")
printLongestSubstringResult("pwwkew")
```
