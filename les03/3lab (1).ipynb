```python
#Создать массив 10x10 со случайными значениями, найти минимум и максимум
import numpy as np


array = np.random.randint(0, 100, size=(10, 10))
print("Исходный массив:")
print(array)

min_value = array.min()
print("\nМинимальное значение:", min_value)


max_value = array.max()
print("Максимальное значение:", max_value)
```

```python
#Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке
import numpy as np

N = 8

matrix = np.zeros((N, N), dtype=int)


for i in range(N):
    for j in range(N):
        if (i + j) % 2 == 0:
            matrix[i][j] = 1
        else:
            matrix[i][j] = 0

print(matrix)
```

```python
#Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8
import numpy as np


z = np.arange(11)


condition = (z >= 3) & (z <= 8)


z[condition] *= -1


print(z)
```

```python
#Найти среднее заначение строки матрицы и вычесть его из каждого елемента строки
import numpy as np


z = np.random.rand(5, 10)


mean_values = np.mean(z, axis=1)

mean_values_reshaped = mean_values.reshape(-1, 1)

result = z - mean_values_reshaped

print(result)
```

```python
#Дан массив 10x2 подеставлющий собой точки в декартовой системе координат, преобразовать в полярную
import numpy as np

z = np.random.random((10, 2))


r = np.sqrt(z[:, 0]**2 + z[:, 1]**2)

theta = np.arctan2(z[:, 1], z[:, 0])

polar_coordinates = np.column_stack((r, theta))

print(polar_coordinates)
```

```python
#Заменить максимальный элемент массива на ноль
import numpy as np


z = np.random.random(10)


idx_max = np.argmax(z)


z[idx_max] = 0

print(z)
```

```python
#Дан массив, добавить 1 к каждому элементу с индексом, заданным в другом массиве. Добавление производиться только 1 раз
import numpy as np


z = np.ones(10)


i = np.random.randint(0, len(z), 20)


unique_indices = np.unique(i)


z[unique_indices] += 1

print(z)
```

```python
#Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением
import numpy as np


z = np.array([1, 2, 3, 4, 5])


new_z = np.repeat(z, 4)


new_z[1::4] = new_z[2::4] = new_z[3::4] = 0

print(new_z)
```

```python
#Найти ближайшее к заданному значению число в заданном массиве
import numpy as np


z = np.arange(100)


v = np.random.uniform(0, 100)


index_closest = np.abs(z-v).argmin()

closest_value = z[index_closest]

print(f"Ближайшее число к {v}: {closest_value}")
```
