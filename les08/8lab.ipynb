```python
#2
import sys
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print("Python version: {sys.version}"),
print("Numpy version: {np.version.version}")
print("Pandas version: {pd.__version__}")
print("Matplotlib version: {mpl.__version__}")
from sklearn.datasets import load_iris
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris_df)
iris_df_scaled = pd.DataFrame(iris_scaled, columns = iris.feature_names)
X = iris_df_scaled
iris_df_scaled.round(2).head()
```

```python
#3
import sys
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print("Python version: {sys.version}"),
print("Numpy version: {np.version.version}")
print("Pandas version: {pd.__version__}")
print("Matplotlib version: {mpl.__version__}")
from sklearn.cluster import DBSCAN
eps = 0.5
min_samples = 5
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
clusters = dbscan.fit_predict(iris_scaled)
iris_df_scaled['Cluster'] = clusters
iris_df_scaled['Cluster'].value_counts()
```

```python
#4
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target
colors = ['red', 'green', 'blue']
labels = iris.target_names

plt.figure(figsize=(8, 6))
for i in range(len(labels)):
    plt.scatter(X[y==i, 0], X[y==i, 1], color=colors[i], label=labels[i])
plt.xlabel('Sepal Length (см)')
plt.ylabel('Sepal Width (см)')
plt.title('Диаграмма рассеяния Sepal Length vs Sepal Width')
plt.legend(title='Виды растений')
plt.grid(True)
plt.show()
```
