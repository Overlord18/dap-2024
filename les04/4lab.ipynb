```python
#Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

#В среднем мужчины 1-го класса старше 40 лет В среднем женщины 1-го класса старше 40 лет Мужчины всех классов в среднем старше женщин того же класса В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
titanic_df = pd.read_csv("1.csv")

# Сначала заполним пропущенные значения средним возрастом
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)

# Средняя возрастная группировка по полу и классу
mean_age = titanic_df.groupby(['Pclass', 'Sex'])['Age'].mean()
print(mean_age)

# Создание подгрупп мужчин и женщин каждого класса
men_first_class = titanic_df[(titanic_df['Pclass'] == 1) & (titanic_df['Sex'] == 'male')]
women_first_class = titanic_df[(titanic_df['Pclass'] == 1) & (titanic_df['Sex'] == 'female')]

men_second_class = titanic_df[(titanic_df['Pclass'] == 2) & (titanic_df['Sex'] == 'male')]
women_second_class = titanic_df[(titanic_df['Pclass'] == 2) & (titanic_df['Sex'] == 'female')]

men_third_class = titanic_df[(titanic_df['Pclass'] == 3) & (titanic_df['Sex'] == 'male')]
women_third_class = titanic_df[(titanic_df['Pclass'] == 3) & (titanic_df['Sex'] == 'female')]

# Расчет средних возрастов мужчин и женщин разных классов
men_first_class_age = round(men_first_class['Age'].mean(), 2)
women_first_class_age = round(women_first_class['Age'].mean(), 2)

men_second_class_age = round(men_second_class['Age'].mean(), 2)
women_second_class_age = round(women_second_class['Age'].mean(), 2)

men_third_class_age = round(men_third_class['Age'].mean(), 2)
women_third_class_age = round(women_third_class['Age'].mean(), 2)

# Анализ утверждений
print("\nАнализ утверждений:")

# Утверждение 1: Среднее число мужчин 1-го класса старше 40 лет?
print("1. В среднем мужчины 1-го класса старше 40 лет:",
      "Верно!" if men_first_class_age > 40 else "Неверно!")

# Утверждение 2: Женщины 1-го класса старше 40 лет?
print("2. В среднем женщины 1-го класса старше 40 лет:",
      "Верно!" if women_first_class_age > 40 else "Неверно!")

# Утверждение 3: Средний возраст мужчин выше женского в каждом классе?
men_older_women = (
    men_first_class_age > women_first_class_age and
    men_second_class_age > women_second_class_age and
    men_third_class_age > women_third_class_age
)
print("3. Мужчины всех классов в среднем старше женщин того же класса:",
      "Верно!" if men_older_women else "Неверно!")

# Утверждение 4: Люди старшего класса старше младшего?
first_class_older_men = men_first_class_age > men_second_class_age > men_third_class_age
first_class_older_women = women_first_class_age > women_second_class_age > women_third_class_age
print("4. В среднем люди в 1-м классе старше, чем во 2-м, а те старше представителей 3-го класса:",
      "Верно!" if first_class_older_men and first_class_older_women else "Неверно!")
```
